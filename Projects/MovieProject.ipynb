# filling of missing **values**

```python
# Calculate the mean of the "Metascore" column
metascore_mean = imdb_df['Metascore'].mean()

# Fill missing values with the mean
imdb_df['Metascore'].fillna(metascore_mean, inplace=True)

```

```python
# Calculate the mode of the "Certificate" column
certificate_mode = imdb_df['Certificate'].mode()[0]

# Fill missing values with the mode
imdb_df['Certificate'].fillna(certificate_mode, inplace=True)
```

```python
#Genre Analysis
genre_counts=imdb_df['Genre'].value_counts()
print(genre_counts)
highest_genre=genre_counts.idxmax()


print("The movie with highest genre is of:",highest_genre)
```

```python
#count the types of Certificate
certificate_counts=imdb_df['Certificate'].value_counts()
print(certificate_counts)
```

```python
#Find the longest and shortest duration movie

longest_duration=imdb_df['Duration'].max()
shortest_duration=imdb_df['Duration'].min()
#print(longest_duration)

longest_duration_movie=imdb_df[imdb_df['Duration']==longest_duration]['Title'].values[0]
shortest_duration_movie=imdb_df[imdb_df['Duration']==shortest_duration]['Title'].values[0]
print("The longest duration movie is:",longest_duration_movie)
print("The shortest duration movie is:",shortest_duration_movie)
```

```python
# which movie performs better audience wise?

highest_rating = imdb_df['Rate'].max()
best_movie = imdb_df.loc[imdb_df['Rate'] == highest_rating, ['Title','Rate']]

# Print the movie with the highest rating
print("Movie which performs  better according to audience is:", best_movie)
```

```python
# which movie performs better gross wise?
highest_gross_votes = imdb_df['Info'].max()
best_movie = imdb_df.loc[imdb_df['Info'] == highest_gross_votes, 'Title'].values[0]

# Print the movie with the highest gross and votes
print("Movie which performs  better according to gross value  is:", best_movie)
```

```python
#Find the movie title with the director Ford Coppola

ford_coppola_movies = imdb_df[imdb_df['Cast'].str.contains("Ford Coppola")]

ford_coppola_movie_titles = ford_coppola_movies['Title']

# Print the movie titles
print("Movies directed by Ford Coppola:")
for title in ford_coppola_movie_titles:
    print(title)

```

```python
#Find the stars of lord of the rings
lord_of_the_rings_stars = imdb_df.loc[imdb_df['Title'].str.contains("The Lord of the Rings"), 'Cast'].values[0]
print("Stars of 'The Lord of the Rings':", lord_of_the_rings_stars)
```

```python
#Find the movies title with the Star Tom Hanks

Tom_Hanks_movies = imdb_df[imdb_df['Cast'].str.contains("Tom Hanks")]

Tom_Hanks_movie_titles = Tom_Hanks_movies['Title']

# Print the movie titles
print("Movies starred by Tom Hanks:")
for title in Tom_Hanks_movie_titles:
    print(title)
```

```python
#Find the top rated movie  with the Star Tom Hanks

tom_hanks_movies = imdb_df[imdb_df['Cast'].str.contains("Tom Hanks")]

# Sort the movies by rating in descending order
sorted_movies = tom_hanks_movies.sort_values('Rate', ascending=False)

# Get the top-rated movie(s) with Tom Hanks
top_rated_movie = sorted_movies.iloc[0]

# Print the top-rated movie with Tom Hanks
print("Top-rated movie with Tom Hanks:")
print(top_rated_movie[['Title', 'Rate']])
```
